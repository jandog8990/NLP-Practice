```python
# Group of texts called a corpus 
# NLTK - contains several corpora from novels in https://www.gutenberg.org/
import nltk
nltk.download("book")
from nltk.book import *
```

```python
# Concordance - see each time a word is used, and immediate context
# how is word being used at sentence level and what words used with it
text8.concordance("man")
```

```python
# occurrences and usages of the word "woman"
text8.concordance("woman")
```

```python
# Disperson plot - view of particular words and how often they occur
# Helps show what words show up near each other. Chronological text.
text8.dispersion_plot(
    ["woman", "lady", "girl", "gal", "man", "gentleman", "boy", "guy"]
)
```

```python
# sense and sensibility - names of houses
#Whitwell appears in the last 3rd of the novel
#Combe appears in first third
text2.dispersion_plot(["Allenham", "Whitwell", "Cleveland", "Combe"])
```

```python
# Frequency dist - which words show up most frequently
from nltk import FreqDist
frequency_distribution = FreqDist(text8)
print(frequency_distribution)
```

```python
# narrow down to the 20 most common words in corpus
frequency_distribution.most_common(20)
freq_dist.plot(20, cumulative=True)
```

```python
# create a list of words that arent stop words
nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
meaningful_words = [
    word for word in text8 if word.casefold() not in stop_words
]
```

```python
# freq dist of non-stop words
freq_dist = FreqDist(meaningful_words)
```

```python
freq_dist.most_common(20)
```

```python
freq_dist.plot(20, cumulative=True)
```

```python
# Collocations - sequence of words that juxtapose each other 
# and that show up often in a particular text
# Examples of collocations that use the word "tree":
# 1. Syntax tree
# 2. Family tree
# 3. Decision tree
text8.collocations()
```

```python
# look for collocations after lemmatizing words.
# find word combos that you missed because now they're lemmas
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in text8]
print(lemmatized_words)
```

```python
# create NLTK text using lemmatized words
new_text = nltk.Text(lemmatized_words)

# create new collocations on lemmatized words
new_text.collocations()

# collocations missing
# 1. weekends away
# 2. poss rship

# new lemmas brought up new collocations
# 1. year old - user often mentions age
# 2. photo pls - user requests one or more photos

# This is how we find common word combos to see what people 
# are talking about and how they're talking about it.
```
